```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

### Load and Explore the Wine Quality Dataset

```python
import pandas as pd
data=pd.read_csv("https://raw.githubusercontent.com/aniruddhachoudhury/Red-Wine-Quality/master/winequality-red.csv")
```

####  Split features and target

```python
X = data.drop('quality',axis=1)
```

```python
X.head()
```

```python
y = data['quality']
```

```python
y.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.33)
```

```python
X_train.head()
```

### Feature scaling

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train) # calculate mean & std deviation
```

```python
print(scaler.mean_)
```

```python
X_train_tf = scaler.transform(X_train)
X_train_tf
```

## Support Vector Classifier (SVC) Model

```python
model = SVC()
```

```python
model.fit(X_train_tf,y_train)
```

```python
# Accuracy
model.score(X_train_tf,y_train)
```

```python
X_test_tf = scaler.transform(X_test)
```

```python

y_predict = model.predict(X_test_tf)
```

```python
y_predict
```

```python
y_test
```

```python
accuracy_score(y_predict,y_test)
```

#### using GridSearchCV

##### Define the Hyperparameter Grid for GridSearchCV

```python
# parameter grid with a range of values to test
param_grid = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto', 0.001, 0.01, 0.1]
}
```

##### Set Up GridSearchCV

```python
# Initialize GridSearchCV
grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    cv=5,  # 5-fold cross-validation
    scoring='accuracy',
    n_jobs=-1,  # Use all available CPU cores
    verbose=1
)
```

```python
# Fit GridSearchCV
grid_search.fit(X_train_tf, y_train)
```

##### Evaluate the Best Model

```python
# Best parameters and score
print("Best Parameters:", grid_search.best_params_)
print("Best Cross-Validation Accuracy:", grid_search.best_score_)
```

```python
# Evaluate on test set
best_model = grid_search.best_estimator_
test_accuracy = best_model.score(X_test_tf, y_test)
print("Test Accuracy:", test_accuracy)
```

## Support Vector Regression (SVR) for ADMISSION_PREDICTION dataset

#### 1. Load and Explore the Dataset

```python
# 1. Load the dataset
url = "https://raw.githubusercontent.com/srinivasav22/Graduate-Admission-Prediction/master/Admission_Predict_Ver1.1.csv"
data = pd.read_csv(url)
data.info();display(data)
```

```python
data.columns = [col.strip() for col in data.columns ] # Stripping the white space from column names
```

```python
data.columns
```

#### 2. Preprocess the data

```python
# Drop Serial No. column
data = data.drop("Serial No.", axis=1)
```

```python
# Check for missing values
data.isnull().sum()
```

#### 3. Split the Dataset

```python
X = data.drop('Chance of Admit',axis=1)
y= data['Chance of Admit']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.33)
```

```python
# Scale features 
# standardize features  using StandardScaler to have zero mean and unit variance.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
print(scaler.mean_)
```

#### 4. Train the SVR Model

```python
# Train SVR model
svr = SVR()
```

```python
svr.fit(X_train_scaled,y_train)
```

```python
svr.score(X_train_scaled,y_train)
```

```python
X_test_transform = scaler.transform(X_test)
```

#### 5.Evaluate the model

```python
y_predict = svr.predict(X_test_transform)
```

```python
mse = mean_squared_error(y_test,y_predict)
rmse = np.sqrt(mse)
r2 = r2_score(y_test,y_predict)
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R² Score: {r2:.4f}")
```

#### 6.Hyperparameter tuning

```python
param_grid = {
    'C': [0.1, 1, 10],
    'epsilon': [0.01, 0.1, 0.5],
    'gamma': ['scale', 'auto', 0.1, 0.01],
    'kernel': ['rbf']
}
grid_search = GridSearchCV(SVR(), param_grid, cv=5, scoring='r2')
grid_search.fit(X_train_scaled, y_train)
print("Best parameters:", grid_search.best_params_)
print("Best R² score:", grid_search.best_score_)
```

#### 7.Predict with best model

```python
best_model = grid_search.best_estimator_
y_pred_best = best_model.predict(X_test_transform)
print(f"Test R² with best model: {r2_score(y_test, y_pred_best):.4f}")
```
